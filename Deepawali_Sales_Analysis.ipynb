```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head(10)
```

```python
df['Status'].unique()
```

```python
df.shape
```

##### rename a column in a DataFrame

```python
df.rename(columns={'Cust_name':'Cust_Name','Age Group':'Age_group'},inplace=True)   
```

# * Data cleaning and Data transformation 

##### # Set the maximum number of rows to be displayed to 800

```python
pd.options.display.max_rows = 100
df.head(100)
```

```python
df.info()
```

##### Dropping columns where all values are NaN i.e., Amount and Status

```python
df.dropna(axis=1, how ='all',inplace=True)
```

```python
df.head()
```

##### Remove unwanted columns 'Product_ID' and 'Age Group'
"Since there are separate columns for age categories, I determined that the columns 'Age Group' hold no significant importance."

```python
df.drop('Age_group',axis=1, inplace=True)
```

```python
df.head()
```

##### converting the 'Amount' column in a pandas DataFrame (presumably named df) to integer data type.
This can be useful when you want to work with the 'Amount' column as integers rather than as floating-point numbers or strings.

```python
# First, replacing NaN and inf values with a suitable default value (e.g., 0)
df['Amount'] = df['Amount'].replace([np.inf, -np.inf, np.nan], 0)

# Then, convert the 'Amount' column to integers
df['Amount'] = df['Amount'].astype('int')
```

```python
df.head(10)
```

##### replacing 0 values with the mean of the amount mean

```python
df['Amount'] = df['Amount'].replace(0, df['Amount'].mean())
```

```python
df.isnull().sum()
```

```python
ax = sns.countplot(x='Gender', data=df)

# Add labels to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=12, color='m', xytext=(0, 5), textcoords='offset points')


ax.set_title('Distribution of Gender')

plt.figure(figsize=(2, 2))   
plt.show()
```

```python
# Group by 'Gender' and sum 'Amount'
gen_df = df.groupby(['Gender'], as_index = False)['Amount'].sum()

sns.barplot(x = 'Gender', y = 'Amount', data = gen_df)
ax = plt.gca()  # Get the current axes
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
state_df = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values('Orders', ascending = False).head(10)
state_df
```

This code groups your DataFrame df by the 'State' column. It creates groups based on the unique values found in the 'State' column.

After grouping by 'State', it calculates the sum of the 'Orders' column within each state group. This is meaningful if 'Orders' represents a numeric variable, and it will give you the total orders for each state.

```python
sns.set(rc={'figure.figsize':(18,5)}) #need to call it before creating the plot.
ax = sns.barplot(x = 'State', y = 'Orders', data = state_df)
```

### 
* In our analysis of order data, we observed significant variations in order volumes across different states. 
* Among the states considered, Uttar Pradesh (UP) exhibited the highest number of orders, followed closely by Maharashtra (MH) and Karnataka (KA). Delhi, Madhya Pradesh (MP), Andhra Pradesh (AP), Himachal Pradesh (HP), Kerala (KL), Haryana (HR), and Gujarat (GJ) also contributed substantially to the order volume, but in descending order.

* Understanding these regional differences in order patterns can be invaluable for our business strategy. 
* It is evident that certain states have a greater demand for our products/services, while others may require more targeted marketing efforts to boost sales. 
* These insights will aid us in tailoring our strategies to optimize our operations and further enhance customer satisfaction.

```python
state_df = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values('Amount',ascending = False).head(10)
state_df
```

```python
state_spending = df.groupby('State')['Amount'].sum().head(5)

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(state_spending, labels=state_spending.index, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Total Amount Spent by State')

# Display the pie chart
plt.show()
```
